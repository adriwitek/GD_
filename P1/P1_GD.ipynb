El objectivo de la practica es tratar de inferir información del profesor ortigosa a partir de un dataset anonimizado.

Comenzamos cargando los datos anonimizados del personal de la UAM del año 2019 obtenido de la siguiente url: https://www.universidata.es/datasets/uam-personal-pdi/recursos/c7309847-b7e1-4b74-85ce-dea0450cb72c#{} Para ello utilizaremos la librería de Python de Pandas, la cula nos falicitara tanto la carga como las consultas necesarias para esta proactoca. Una vez cargado el dataset, comprobamos su estructura comprobando que se compone de 164928 registros distribuidos en 2577 tuplas con 64 campos cada una.

```python
import pandas as pd

path = "uam-personal-pdi-2019-anonimizado.csv";
df = pd.read_csv (path)

print("Tamaño del dataset: ", df.size)
print("Forma del dataset: ", df.shape)
```

Visualizamos un pequeño fragmento del datset para visualizarlo mejor:

```python
df.head(10)
```

Tal como dice el enunciado, comenzamos descartando aquellos campos cuyo nombre empieza por "cod_" "lat_" y "lon_". Tras eso volvemos a calcular las dimensiones del dataset, el cual se ha reducido a 90195 registros distribuidos en 2577 tuplas previas  y reducido a tan solo 35 campos.

```python
df = df.loc[:,~df.columns.str.startswith('cod_')]
df = df.loc[:,~df.columns.str.startswith('lat_')]
df = df.loc[:,~df.columns.str.startswith('lon_')]
df = df.loc[:,~df.columns.str.startswith('IMPORTANTE')]

print("Tamaño del dataset: ", df.size)
print("Forma del dataset: ", df.shape)
```

Volvemos a sacar una pequeña muestra:

```python
df.head(10)
```

En nuestro caso los campos que actuarán como variables pivote serán: des_unidad_responsable y des_genero, afortunadamente para el desarrollo de esta práctica disponemos de cierta información dada en el enunciado:

<ul>
  <li>El profesor Ortigosa estaba en 2019 en la UAM (=aparece en el fichero)</li>
  <li>El profesor Ortigosa era de género masculino en 2019</li>
  <li>El profesor Ortigosa pertenecía al Departamento de Ingeniería Informática en 2019</li>
</ul>

Tanto el genero como la unidad responsable son datos que conforman las variables pivotes del dataset, por lo que si hacemos un filtrado con estos parámetros podemos reducir el dataset a un único grupo que deberá contener la información de dataset. Por lo que procedemos a ello.

```python
aux = df[df["des_genero"].str.contains('Hombre') & 
         df["des_unidad_responsable"].str.contains('Departamento de Ingeniería Informática')]

print("Tamaño del dataset: ", aux.size)
print("Forma del dataset: ", aux.shape)
```

Con este filtro hemos logrado acotar el grupo que contiene al profesor Ortigosa, cuya información estará almacenada en una de las 56 tuplas que tenemos. Procedemos a sacar un fragmento del grupo del dataset que contiene la info del profesor.

```python
aux.head(5)
```

Ahora procedemos a ir bloque por bloque, como las tuplas de cada bloque estarán permutadas necesitaremos al menos información de uno de los campos de cada bloque para extraer información, es decir, necesitaremos un dato externo por cada bloque para poder inferir información con cierta probabilidad.

Para obtener esta información extenra vamos a utilizar las siguientes fuentes:

</ul>

Nota: Los bloques que solo tiene un campo no tiene sentido hacer inferencia, ya que no obtendriamos nueva información no obtenida ya por medios externos. Por lo que realizaremos otro tipo de análisis dependiendo del valor.
